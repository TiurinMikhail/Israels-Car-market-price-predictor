```python
from deep_translator import GoogleTranslator

translator = GoogleTranslator(source='auto', target='en')

# Function to translate text using deep_translator
def deep_translate_text(text, target='en'):
    return translator.translate(text)

# Apply the translation function to the DataFrame
df_fin['car_name_en'] = df_fin['car_name'].apply(deep_translate_text)

df_fin
```

```python
df_fin['car_info_en'] = df_fin['car_info'].apply(deep_translate_text)
```

```python
df_fin['car_tags_en'] = df_fin['car_tag'].apply(lambda x: deep_translate_text(x) if pd.notna(x) else x)
```
