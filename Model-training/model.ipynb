```python
mercedes_gl = df_yad2.loc[(df_yad2['brand'] == 'Mercedes-Benz') & (df_yad2['model'] == 'GL')]

import re 

def replace_model(row):
    if row['brand'] == 'Mercedes-Benz' and row['model'] == 'GL':
        model = re.findall('[A-Z][A-Z][A-Z]', row['car_name_en'])
        if model:
            return model[0]
    return row['model']

mercedes_gl['model'] = df_yad2.apply(replace_model, axis=1)
mercedes_gl

df_yad2.loc[mercedes_gl.index, 'model'] = mercedes_gl['model']
```

```python
from collections import Counter

model_counts = Counter(df_yad2['model'])
most_popular_model = model_counts.most_common(1)[0][0]
print(f"The most popular model in the dataset is: {most_popular_model}")
```

```python
import matplotlib.pyplot as plt

model_counts = Counter(df_yad2['model'])
top_10_models = model_counts.most_common(10)

models = [model for model, count in top_10_models]
counts = [count for model, count in top_10_models]

plt.figure(figsize=(10, 6))  # increase the figure size for better visibility
plt.bar(models, counts, color=plt.cm.tab20(range(len(models))))  # use different colors for different bars
plt.xlabel('Model', fontsize=14)  # increase the font size of the x-axis label
plt.ylabel('Number of listed cars', fontsize=14)  # increase the font size of the y-axis label
plt.title('Top 10 Most Popular Models', fontsize=16)  # increase the font size of the title
plt.show()
```

```python
import matplotlib.pyplot as plt

brand_counts = Counter(df_yad2['brand'])
top_10_brands = brand_counts.most_common(10)

brands = [brand for brand, count in top_10_brands]
counts = [count for brand, count in top_10_brands]

plt.figure(figsize=(10, 6))  # increase the figure size for better visibility
plt.bar(brands, counts, color=plt.cm.tab20(range(len(brands))))  # use different colors for different bars
plt.xlabel('Brand', fontsize=14)  # increase the font size of the x-axis label
plt.ylabel('Number of listed cars by brand', fontsize=14)  # increase the font size of the y-axis label
plt.title('Top 10 Most Popular Brands', fontsize=16)  # increase the font size of the title
plt.show()
```

```python
import matplotlib.pyplot as plt

hand_num_counts = Counter(df_yad2['hand_num'])
hand_num_distribution = hand_num_counts.most_common()

hand_nums = [hand_num for hand_num, count in hand_num_distribution]
counts = [count for hand_num, count in hand_num_distribution]

plt.figure(figsize=(10, 6))  
plt.bar(hand_nums, counts, color=plt.cm.tab20(range(len(hand_nums))))  
plt.xlabel('Hand Number', fontsize=14)  
plt.ylabel('Number of listed cars', fontsize=14)  
plt.title('Distribution of Hand Numbers', fontsize=16)  
plt.show()
```

```python
import matplotlib.pyplot as plt

average_price = df_yad2.groupby('car_name_en')['price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))  
plt.bar(range(len(average_price)), average_price.values, color=plt.cm.tab20(range(len(average_price))))  
plt.xticks(range(len(average_price)), average_price.index, rotation=45, ha='right')  
plt.xlabel('Car Name', fontsize=14)  
plt.ylabel('Average Price', fontsize=14)  
plt.title('Top 10 Most Expensive Cars by Average Price', fontsize=16)  
plt.tight_layout()  
plt.show()
```
